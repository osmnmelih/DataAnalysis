This project analyzes student math practice data to uncover learning patterns, performance trends, and areas for instructional improvement.

## 1Ô∏è‚É£ Setup & Data Loading

```

import pandas as pd

df = pd.read_csv("math_practice_data.csv")
df.head()
```

## 2Ô∏è‚É£ Exploratory Data Analysis

```

df.describe()
```

## 3Ô∏è‚É£ Performance by Operation

```

operation_stats = (
    df.groupby('operation')
    .agg(
        total_problems=('problem', 'count'),
        avg_time=('time_spent_seconds', 'mean'),
        success_rate=('is_correct', 'mean'),
        avg_attempts=('attempts_needed', 'mean')
    )
)

operation_stats['success_rate'] *= 100
operation_stats
```

## 4Ô∏è‚É£ Performance by Difficulty

```

difficulty_stats = (
    df.groupby('difficulty_level')
    .agg(
        total_problems=('problem', 'count'),
        avg_time=('time_spent_seconds', 'mean'),
        success_rate=('is_correct', 'mean'),
        avg_attempts=('attempts_needed', 'mean')
    )
)

difficulty_stats['success_rate'] *= 100
difficulty_stats
```

## 5Ô∏è‚É£ Student Performance Report Cards

```

student_stats = (
    df.groupby('student_id')
    .agg(
        total_attempts=('problem', 'count'),
        success_rate=('is_correct', 'mean'),
        avg_time=('time_spent_seconds', 'mean'),
        avg_attempts=('attempts_needed', 'mean')
    )
    .sort_values(by='success_rate', ascending=False)
)

student_stats['success_rate'] *= 100
student_stats
```

## 6Ô∏è‚É£ Correlation Analysis

```

df[['time_spent_seconds', 'attempts_needed', 'is_correct']].corr()
```

## 7Ô∏è‚É£ Learning Progress Over Time

```

session_performance = df.groupby('session_number')['is_correct'].mean() * 100
session_performance
```

## 8Ô∏è‚É£ Identifying At-Risk Students

```

at_risk = student_stats[
    (student_stats['success_rate'] < 60) &
    (student_stats['avg_attempts'] > 2)
]
at_risk
```

## üß† Key Insights

- Students improve significantly over sessions
- Hard problems and multiplication require targeted intervention
- Attempts and time correlate negatively with accuracy
- Student-level analytics enable early intervention
